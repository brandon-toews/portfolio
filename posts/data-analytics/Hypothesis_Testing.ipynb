```python
# Import dependencies
import pandas as pd
import numpy as np
from scipy import stats

#import cleaned dataset
df = pd.read_parquet('cleaned_dataset.parquet')

# split df into two 
#dataframe with entries just for the fixed category
df_fixed = df.loc[df['net_type'] == 'Fixed'].copy()
df_fixed.reset_index(drop=True, inplace=True)
df_fixed.drop(columns='net_type', inplace=True)

#dataframe with entries just for the mobile category
df_mob = df.loc[df['net_type'] == 'Mobile'].copy()
df_mob.reset_index(drop=True, inplace=True)
df_mob.drop(columns='net_type', inplace=True)

display(df)
print(df.columns)

# create dataframe dict to iterate through
dfs = {
    'Fixed': df_fixed,
    'Mobile': df_mob,
    # import YeoJ transformed dfs
    'YeoJ_Fix': pd.read_parquet('yeoj_fixed_dataset.parquet'),
    'YeoJ_Mob': pd.read_parquet('yeoj_mob_dataset.parquet')
}

for key in dfs:
    display(dfs[key])
    print(dfs[key].columns)
    
```

## Levene's test

```python
#| label: lev-test

# Extract 'avg_d_mbps' for the fixed and mobile networks
fixed_avg_d_mbps = dfs['Fixed']['avg_d_mbps']
mobile_avg_d_mbps = dfs['Mobile']['avg_d_mbps']

# Perform the Levene's test
statistic, p_value = stats.levene(fixed_avg_d_mbps, mobile_avg_d_mbps)

# Display results
print('Statistic:', statistic)
print('p-value:', np.round(p_value, 10))
```

# F-Tests

First, run an F-test on the untransformed positively skew data 

```python
# Extract 'avg_d_mbps' for the fixed and mobile networks of the yeo-j transformed
fixed_avg_d_mbps = dfs['Fixed']['avg_d_mbps']
mobile_avg_d_mbps = dfs['Mobile']['avg_d_mbps']

# Calculate the sample variances for fixed and mobile
fixed_variance = np.var(fixed_avg_d_mbps, ddof=1)
mobile_variance = np.var(mobile_avg_d_mbps, ddof=1)

# Perform the F-test
f_statistic = fixed_variance / mobile_variance
df1 = len(fixed_avg_d_mbps) - 1
df2 = len(mobile_avg_d_mbps) - 1

# Calculate two-tailed p-value by calculating the CDF of the
# F-distribution at the f_statistic
p_value = 2 * (1 - stats.f.cdf(np.abs(f_statistic), df1, df2))

# Display results
print('F statistic:', f_statistic)
print('p-value:', np.round(p_value, 10))

# Print which network has larger variance
if f_statistic > 1:
    print("The fixed network has a larger variance.")
elif f_statistic < 1:
    print("The mobile network has a larger variance.")
else:
    print("The variances are equal.")
```

Second, run an F-test on the Yeo-Johnson transformed data 

```python
#| label: f-test-yeoj

# Extract 'avg_d_mbps' for the fixed and mobile networks of the yeo-j transformed
fixed_avg_d_mbps = dfs['YeoJ_Fix']['avg_d_mbps']
mobile_avg_d_mbps = dfs['YeoJ_Mob']['avg_d_mbps']

# Calculate the sample variances for fixed and mobile
fixed_variance = np.var(fixed_avg_d_mbps, ddof=1)
mobile_variance = np.var(mobile_avg_d_mbps, ddof=1)

# Perform the F-test
f_statistic = fixed_variance / mobile_variance
df1 = len(fixed_avg_d_mbps) - 1
df2 = len(mobile_avg_d_mbps) - 1

# Calculate two-tailed p-value by calculating the CDF of the
# F-distribution at the f_statistic
p_value = 2 * (1 - stats.f.cdf(np.abs(f_statistic), df1, df2))

# Display results
print('F statistic:', f_statistic)
print('p-value:', np.round(p_value, 10))

# Print which network has larger variance
if f_statistic > 1:
    print("The fixed network has a larger variance.")
elif f_statistic < 1:
    print("The mobile network has a larger variance.")
else:
    print("The variances are equal.")
```

## T-Tests

Firt, run normal skewed data.


```python
# Extract 'avg_d_mbps' for the fixed and mobile networks
fixed_avg_d_mbps = dfs['Fixed']['avg_d_mbps']
mobile_avg_d_mbps = dfs['Mobile']['avg_d_mbps']

# Perform the t-test
t_statistic, p_value = stats.ttest_ind(fixed_avg_d_mbps, mobile_avg_d_mbps)

# Check the sign of the t-statistic
# Print which network has larger average download speed
if t_statistic > 0:
    print("The fixed network has a larger average download speed.")
elif t_statistic < 0:
    print("The mobile network has a larger average download speed.")
else:
    print("The average download speeds are equal.")

# Display results
print('t statistic:', t_statistic)
print('p-value:', p_value)
```

Second, run t-test with Yeo-J transformed data

```python
# Extract 'avg_d_mbps' for the fixed and mobile networks
fixed_avg_d_mbps = dfs['YeoJ_Fix']['avg_d_mbps']
mobile_avg_d_mbps = dfs['YeoJ_Mob']['avg_d_mbps']

# Perform the t-test
t_statistic, p_value = stats.ttest_ind(fixed_avg_d_mbps, mobile_avg_d_mbps)

# Check the sign of the t-statistic
# Print which network has larger average download speed
if t_statistic > 0:
    print("The fixed network has a larger average download speed.")
elif t_statistic < 0:
    print("The mobile network has a larger average download speed.")
else:
    print("The average download speeds are equal.")

# Display results
print('t statistic:', t_statistic)
print('p-value:', p_value)
```

## Man-Whitney U Test

Run the untransformed data of this non-parametric test

```python
# Extract 'avg_d_mbps' for the fixed and mobile networks
fixed_avg_d_mbps = dfs['Fixed']['avg_d_mbps']
mobile_avg_d_mbps = dfs['Mobile']['avg_d_mbps']

# Perform the Mann-Whitney U test
u_statistic, p_value = stats.mannwhitneyu(fixed_avg_d_mbps, mobile_avg_d_mbps, alternative='two-sided')

# Check the medians of the two groups
# whichever median is higher tends to have larger speeds
if np.median(fixed_avg_d_mbps) > np.median(mobile_avg_d_mbps):
    print("The fixed network tends to have larger download speeds.")
elif np.median(fixed_avg_d_mbps) < np.median(mobile_avg_d_mbps):
    print("The mobile network tends to have larger download speeds.")
else:
    print("The download speeds are similar for both networks.")

# Display results
print('U statistic:', u_statistic)
print('p-value:', p_value)
```
